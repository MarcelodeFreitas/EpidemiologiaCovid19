```python
hostname = 'localhost'
username = 'postgres'
password = '1234'
database = 'Miebiom'
```

```python
def doQuery( conn ) :
    cur = conn.cursor()

    cur.execute( "SELECT concelho FROM concelho" )

    for concelho in cur.fetchall() :
        print (concelho)

myConnection = psycopg2.connect( host=hostname, user=username, password=password, dbname=database )
doQuery( myConnection )
```
